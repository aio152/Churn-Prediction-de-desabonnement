```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
X_data = np.random.random(50)*100
y_data = np.random.random(50)*100

plt.scatter(X_data,y_data, c= "blue", marker="*", s = 150, alpha = 0.1)
plt.show()
```

```python
years = [2006 + x for x in range(12)]

weights = [80, 83, 84, 86, 82, 70, 76, 81, 85, 86, 84, 85]

plt.plot(years, weights, c="m", lw = 10, linestyle = "--")
plt.show()
```

```python
x = ["C++", "C#", "python", "Java", "Go"]
y = [20, 50, 140, 1, 45]

plt.bar(x,y, color = "black" , align = "center", width = 0.5, edgecolor = "green", lw = 3)
```

```python
ages =np.random.normal(20, 1.5, 1000)

plt.hist(ages, bins =60, color = "red")
```

```python
langs = ["C++", "C#", "python", "Java", "Go"]
votes = [20, 50, 140, 10, 45]
explodes = [0.7, 0.1, 0, 0, 0]
plt.pie(votes,labels = langs, autopct = "%.2f%%", pctdistance = 0.7, startangle = 90)
plt.show()
```
